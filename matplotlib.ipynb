```python
## matplotlib -----------------------

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5,7,8]

y = [10,12,14,16,15,17,19]

```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y , label = 'line plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
plt.plot(x,y , label = 'line plot' , color = 'red' , linewidth = 2 , linestyle = '-')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
plt.plot(x,y , label = 'line plot' , color = 'c' , linewidth = 1 , linestyle = '--' , marker = 'o', markersize = 7)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15, 7))
plt.plot(x,y , label = 'line plot' , color = 'y' , linewidth = 1 , linestyle = '--' , marker = 'o', markersize = 7)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15, 7))
plt.plot(x,y , label = 'line plot' , color = 'g' , linewidth = 2 , linestyle = ':' , marker = 'o', markersize = 7)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot')
plt.legend()
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]

y1 = [8,5,9,11,7]

y2 = [9,7,5,6,10]
```

```python
plt.figure(figsize = (15, 7))
plt.plot(x,y1 , label = 'line plot X Vs Y1' , color = 'r' , linewidth = 1 , linestyle = ':' , marker = '*', markersize = 7)
plt.plot(x,y2 , label = 'line plot X Vs Y2' , color = 'c' , linewidth = 1 , linestyle = ':' , marker = '^', markersize = 7)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Multiple Line Plot')
plt.legend()
plt.grid()
plt.show()
```

```python
## Bar plot -----------------------------------

class_name = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

num_of_class = [10, 20, 30, 40, 50, 60, 70, 80]

```

```python
plt.bar(class_name , num_of_class)
```

```python
plt.figure(figsize = (15, 7))
plt.bar(class_name , num_of_class , color = 'green' , width=0.5   )
plt.xlabel('class_name')
plt.ylabel('num_of_class')
plt.title('Bar Plot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15, 7))
plt.barh(class_name , num_of_class , color = 'c' , height=0.5  )
plt.xlabel('class_name')
plt.ylabel('num_of_class')
plt.title('Bar Plot')
plt.legend()
plt.show()
```

*** ## histogram plot ------------------------***

```python
## histogram plot ------------------------

data = np.random.normal(0 ,1 , 1000)


plt.figure(figsize = (15, 7))
plt.hist(data, color = 'r' , linewidth = 1, edgecolor = 'k',  )
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Plot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15, 7))
plt.hist(data, color = 'r' , linewidth = 1, edgecolor = 'k', bins = 20, alpha = 0.7, label = 'Histogram'  )
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Plot')
plt.legend()
plt.show()
```

```python

data = np.random.normal(0 ,1 , 1000)
data1 = np.random.normal(3 ,1 , 1000)


plt.figure(figsize = (15, 7))
plt.hist([data ,data1], color = ['r' , 'b'] , linewidth = 1, edgecolor = 'k',  )
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Plot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15, 7))
plt.hist([data ,data1], color = ['r' , 'b'] , linewidth = 1, edgecolor = 'k', label=['data' , 'data1']  )
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Plot')
plt.legend()
plt.show()
```

```python
data = np.random.gamma(2,1,400)


plt.figure(figsize = (15, 7))
plt.hist(data, color = 'c' , linewidth = 1, edgecolor = 'k', label= 'data', width=0.5 , alpha=0.5 )
plt.axvline(data.mean(), color = 'r', linestyle = '--', label = 'mean')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Plot')
plt.legend()
plt.show()
```

```python
class_name = ['Python' , 'C++' , 'Java' , 'Deep Learning']

class_size = [75 ,85,92,90]
```

```python
plt.pie( class_size )
plt.show()
```

```python
plt.pie( class_size , labels=class_name )
plt.show()
```

```python
plt.pie( class_size , labels=class_name , autopct='%1.1f%%' , shadow=True , startangle=90)
plt.show()
```

```python
explode = [0.02 , 0 , 0.1 , 0]

plt.pie( class_size , labels=class_name , autopct='%1.1f%%' , shadow=True , startangle=90 , explode= explode)
plt.show()
```
